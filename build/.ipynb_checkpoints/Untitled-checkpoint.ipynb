```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

hits = [
3.4066,
5.42278,
11.2194,
19.2291,
23.3919,
24.3479,
25.5392,
30.9642,
33.0784,
48.8187
]
realhits = [
    3.4066,
    19.2291,
    
]
time, amplitude = np.loadtxt("WavesFORTHESIS",usecols=(0,1),unpack=True,comments=":")

for i in hits:
    plt.axvline(i, linewidth = 0.5)
plt.axhline(7, linewidth = 0.4)
plt.ylim(0,50)
plt.plot(time,amplitude, '.', markersize= 3)
plt.show()
```
