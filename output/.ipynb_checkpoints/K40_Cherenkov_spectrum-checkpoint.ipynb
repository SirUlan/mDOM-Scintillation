```python
wavelengths = np.loadtxt(folder+"ch_spectr_vitrovex_spice.txt", unpack=True)
print len(wavelengths)
wavelengths2 = np.loadtxt(folder+"ch_spectr_vitrovex_air.txt", unpack=True)
print len(wavelengths2)
```

```python
cols = sns.color_palette("deep", 10)
fig = plt.figure(figsize=(10,8))
ax = fig.gca()

binwidth = 10

plt.hist(wavelengths, bins=range(100, 800 + binwidth, binwidth), label="orig spectr spice")
plt.hist(wavelengths2, bins=range(100, 800 + binwidth, binwidth), label="orig spectr air")


ax.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel("Cherenkov photon wavelength [nm]", fontsize=18)
plt.ylabel("hits", fontsize=18)
plt.grid(True)
#plt.xlim(0, 180)
#plt.ylim(0, 1.1)
plt.legend(loc="best", fontsize=15)
```
