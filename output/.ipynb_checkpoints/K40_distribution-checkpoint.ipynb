## K40 hit properties

```python
# 789 = xyz
# event#  hit time/ns  flight time/ns  track length/m  energy/eV PMT#  event distance/m  photon position[m] x,y,z  direction x,y,z distance[m]
#num_event, track_length, num_pmt, event_dist, hit_x, hit_y, hit_z  = np.loadtxt("test_scan_normalDOM_corr_t0_individual_n10000000.txt", usecols=(0,3,5,6,7,8,9), unpack=True)
num_event, track_length, energy_eV, num_pmt, event_dist, hit_x, hit_y, hit_z  = np.loadtxt("mdom_k40_data_1e6.txt", usecols=(0,3,4,5,6,7,8,9), unpack=True)
```

```python
sns.set_palette("deep")
print "Total hits", len(num_event)

nums = [0, 1, 2, 3]
for num in nums:
    print "Hits on PMT #", num, len(num_event[num_pmt==num])

fig = plt.figure(figsize=(15,12))
ax = fig.gca()

#condition = [num_pmt==0]
#condition = [(num_pmt==0) | (num_pmt==1)]
condition = [(num_pmt>=12) & (num_pmt<=23)]

# #norm=LogNorm()
plt.hist2d(hit_x[condition], hit_y[condition], bins=400)
plt.colorbar()


# plt.legend(loc="best", fontsize=15)
# plt.xlim(-1, 181)
# plt.grid(True)

# ax.set_xlabel("theta", fontsize=20)
# ax.set_ylabel("detected/emitted", fontsize=20)
# ax.tick_params(axis='both', which='major', labelsize=15)
```

```python
cols = sns.color_palette("deep", 10)
fig = plt.figure(figsize=(10,8))
ax = fig.gca()

plt.hist(energy_eV, bins=50, label="spectrum @ PMTs, KM3NeT")

ax.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel("photon energy [eV]", fontsize=18)
plt.ylabel("hits", fontsize=18)
plt.grid(True)
#plt.xlim(0, 180)
#plt.ylim(0, 1.1)
plt.legend(loc="best", fontsize=15)
```

```python
fig = plt.figure(figsize=(15,12))
ax = fig.gca()

binwidth = 10
plt.hist(1240/energy_eV, bins=range(100, 800 + binwidth, binwidth), label="spectrum @ PMTs, mDOM", color="orange")

ax.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel("wavelength [nm]", fontsize=18)
plt.ylabel("hits", fontsize=18)
plt.grid(True)
plt.xlim(100, 800)
#plt.ylim(0, 1.1)
plt.legend(loc="best", fontsize=15)
```
